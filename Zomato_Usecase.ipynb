# Encode the Remaining columns

```python
def encode(zomato):
    for column in zomato.columns[~zomato.columns.isin(['rate','vote','cost'])]:
        zomato[column] = zomato[column].factorize()[0]
    return zomato
```

```python
zomato4 = encode(zomato3)
```
